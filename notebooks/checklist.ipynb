## Kiểm tra kết nối LLM (The Brain)

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="qwen2.5:3b")
try:
    response = llm.invoke("Xin chào, bạn có thể giúp tôi không?")
    print("Kết nối Ollama: THÀNH CÔNG")
    print("Phản hồi:", response.content)
except Exception as e:
    print("Lỗi kết nối Ollama:", e)
```

```python
import os
from langchain_nvidia_ai_endpoints import ChatNVIDIA
from dotenv import load_dotenv

load_dotenv()

# Khởi tạo Llama 3.1 70B - Một trong những model mạnh nhất hiện nay
# Model này vượt xa Qwen 3B về khả năng lập luận (Reasoning)
llm = ChatNVIDIA(
    model="meta/llama-3.1-70b-instruct",
    api_key=os.getenv("NVIDIA_NIM_API_KEY"),
    temperature=0.2,
    max_completion_tokens=1024
)

print("Kết nối NVIDIA Llama 3.1 70B: THÀNH CÔNG", llm.model)
```

## Kiểm tra khả năng truy cập Web (The Senses)

```python
from ddgs import DDGS

with DDGS() as ddgs:
    results = [r for r in ddgs.text("thị trường smartphone Việt Nam 2026", max_results=3)]
    if results:
        print("Kết nối DuckDuckGo: THÀNH CÔNG")
    else:
        print("Không tìm thấy dữ liệu web.")
```
